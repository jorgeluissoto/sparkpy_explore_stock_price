### Explore and Query Data
##### DataFrames operations

```python
from pyspark.sql.functions import year, month
```

```python
disney_df.select(year("Date").alias("yr"), "Close")
  .groupby("yr").avg("Close").sort("yr").show()
```

```python
disney_df.select(year("Date").alias("year"),
                month("Date").alias("month"),
                "Low")\
          .groupby("year", "month") \
          .avg("Low") \
          .sort("year", "month") \
          .show()
```

```python
disney_df.select(year("Date").alias("year"),
                month("Date").alias("month"),
                "Low")\
          .groupby("year", "month") \
          .avg("Low") \
          .sort("year", "month") \
          .explain()
```
